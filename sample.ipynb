### if文, else文, elif文

```python
score = 100
if score >= 60:
    print("合格です。")
```

```python
score = 100
if score == 100:
    print("満点です。")
```

```python
score = 100
if score = 100:
    print("満点です。")
```

```python
score = 99
if score != 100:
    print("満点ではありません。")
```

```python
n = 10
if n % 2 == 0:
    print("偶数です")
```

```python
password = "abc123"
if password == "abc123":
    print("認証成功です。")
```

```python
password = "xyz789"
if password != "abc123":
    print("認証失敗です。")
```

```python
score = 20
if score >= 60:
    print("合格です。")
else:
    print("不合格です。")
```

```python
score = 80
result = "合格" if score >= 60 else "不合格"
print(result + "です。")
```

```python
score = 70
if score >= 80:
    print("優")
elif score >= 65:
    print("良")
elif score >= 60:
    print("可")
else:
    print("不可")
```

### 論理演算子

```python
print(True)
print(not True)
```

```python
score = 40
if not score >= 60:
    print("不合格です。")
```

```python
print(False and False)
print(True  and False)
print(False and True)
print(True  and True)
```

```python
has_license = False
is_driving = True

if not has_license and is_driving:
    print("⚠️ 無免許運転です！")
```

```python
print(False or False)
print(True  or False)
print(False or True)
print(True  or True)
```

```python
is_student = True
is_senior = False

if is_student or is_senior:
    print("割引料金になります。")
```
